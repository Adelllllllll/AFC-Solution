# 2. Vérification des dates

```python
# On tente une conversion pour voir si ça plante
try:
    df['sale_date_clean'] = pd.to_datetime(df['sale_date'])
    print("\n--- 3. Dates converties avec succès ---")
    print(df['sale_date_clean'].head())
except Exception as e:
    print(f"\nERREUR DATE : {e}")
```

# 3. Vérification de la cohérence des montants

```python
# On vérifie si Total = Prix * Quantité (à une petite marge d'erreur près pour les float)
df['calcul_theorique'] = df['quantity'] * df['unit_price']
df['ecart'] = df['total_amount'] - df['calcul_theorique']

# On affiche les lignes où l'écart est significatif (> 0.01)
incoherences = df[abs(df['ecart']) > 0.01]

if not incoherences.empty:
    print(f"\n--- 4. ATTENTION : {len(incoherences)} lignes avec incohérence de prix ---")
    display(incoherences)
else:
    print("\n--- 4. Calculs cohérents (Total = Qte * Prix) ---")
```
